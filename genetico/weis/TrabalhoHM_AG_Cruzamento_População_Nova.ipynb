```python
import numpy as np
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
import random
import matplotlib.pyplot as plt
def populacao_incial():
    list_ind_crescente_prob = pd.read_excel('População_inicial .xlsx')
    individuo = list(range(1,51))
    list_ind_crescente_prob= list_ind_crescente_prob.set_axis(individuo, axis='index')
    list_ind_crescente_prob = pd.DataFrame(list_ind_crescente_prob)
    jobs = 10
    machines = 2
    jm = jobs + machines
    j=1
    m11 = []
    for j in range (1,51):
        m1 = list_ind_crescente_prob.loc[j, 'm1']
        m1 = [int(i) for i in m1.split() if i.isdigit()]
        m1 = list([m1])
        m11.append(m1)
        j= j +1
    j=1
    m22 = []
    for j in range (1,51):
        m2 = list_ind_crescente_prob.loc[j, 'm2']
        m2 = [int(i) for i in m2.split() if i.isdigit()]
        m2 = list([m2])
        m22.append(m2)
        j= j +1
    m11 = pd.DataFrame(m11)
    m11 = m11.set_axis(individuo, axis='index')
    m1_n = list(['m1'])
    m11 = m11.set_axis(m1_n, axis='columns')
    list_ind_crescente_prob['m1'] = m11
    m22 = pd.DataFrame(m22)
    m22 = m22.set_axis(individuo, axis='index')
    m2_n = list(['m2'])
    m22 = m22.set_axis(m2_n, axis='columns')
    list_ind_crescente_prob['m2'] = m22
    #print (list_ind_crescente_prob)
    return list_ind_crescente_prob
```

```python
# Selecionar grupos e pesos = roleta
# População separada em 5 partes
def roleta(list_ind_crescente_prob):
    gera = 0
    list_filho1_m1 = []
    list_filho1_m2 = []
    list_make_filho1 = []
    list_filho2_m1 = []
    list_filho2_m2 = []
    list_make_filho2 = []
    for gera in range (25):
        peso_pop1 = np.sum(list_ind_crescente_prob.loc[1:10,'Probabilidade'])
        peso_pop2 = np.sum(list_ind_crescente_prob.loc[11:20,'Probabilidade']) 
        peso_pop3 = np.sum(list_ind_crescente_prob.loc[21:30,'Probabilidade']) 
        peso_pop4 = np.sum(list_ind_crescente_prob.loc[31:40,'Probabilidade']) 
        peso_pop5 = np.sum(list_ind_crescente_prob.loc[41:50,'Probabilidade']) 
        list_peso = [peso_pop1, peso_pop2, peso_pop3, peso_pop4, peso_pop5]
        pop1 = individuo[0:10]
        pop2 = individuo[10:20]
        pop3 = individuo[20:30]
        pop4 = individuo[30:40]
        pop5 = individuo[40:50]
        list_pop = [pop1, pop2, pop3, pop4, pop5]
        list_pop_rad = random.choices(list_pop, weights=list_peso, k=2)
        pai1_loc = random.choice(list_pop_rad[0])
        pai2_loc = random.choice(list_pop_rad[1])
        pai1 = list_ind_crescente_prob.loc[pai1_loc]
        pai2 = list_ind_crescente_prob.loc[pai2_loc]
        pai1_m1 = pai1[0]
        pai1_m2 = pai1[1]
        apt_pai1 = pai1[2]
        pai2_m1 = pai2[0]
        pai2_m2 = pai2[1]
        apt_pai2 = pai2[2]
        #Crossover one point
        # Escolha aleatória de ponto de corte em cada máquina o pai1
        ponto_pai1_m1 = np.int32(random.choice(pai1_m1))
        ponto_pai1_m2 = np.int32(random.choice(pai1_m2))
        #print('-----------------Seg1 - m1-------------------------')
        loc_ponto_pai1_m1 = np.argwhere(pai1_m1 == ponto_pai1_m1)
        loc_ponto_pai1_m1 = loc_ponto_pai1_m1[0]
        loc_ponto_pai1_m1 = loc_ponto_pai1_m1[0]
        seg1_ponto_pai1_m1 = pai1_m1[0:loc_ponto_pai1_m1+1]
        #print ('---------------Seg2 - m1--------------------------')
        loc_ponto_pai1_m1 = np.argwhere(pai1_m1 == ponto_pai1_m1)
        loc_ponto_pai1_m1 = loc_ponto_pai1_m1[0]
        loc_ponto_pai1_m1 = loc_ponto_pai1_m1[0]
        seg2_ponto_pai1_m1 = pai1_m1[loc_ponto_pai1_m1+1:len(pai1_m1)]
        #print ('---------------Seg1 - m2--------------------------')
        #Pai 1 - Segmento 1 da m2
        loc_ponto_pai1_m2 = np.argwhere(pai1_m2 == ponto_pai1_m2)
        loc_ponto_pai1_m2 = loc_ponto_pai1_m2[0]
        loc_ponto_pai1_m2 = loc_ponto_pai1_m2[0]
        seg1_ponto_pai1_m2 = pai1_m2[0:loc_ponto_pai1_m2+1]
        #print ('------------------Seg2 - m2-----------------------')
        loc_ponto_pai1_m2 = np.argwhere(pai1_m2 == ponto_pai1_m2)
        loc_ponto_pai1_m2 = loc_ponto_pai1_m2[0]
        loc_ponto_pai1_m2 = loc_ponto_pai1_m2[0]
        seg2_ponto_pai1_m2 = pai1_m2[loc_ponto_pai1_m2+1:len(pai1_m2)]
        ## Filho1 M1
        pai2 = []
        pai2.extend(pai2_m1)
        pai2.extend(pai2_m2)
        #print('_______________Pai 2______________________')
        filho1 = []
        filho1.extend(seg1_ponto_pai1_m1)
        filho1.extend(seg1_ponto_pai1_m2)
        #print('_______________Filho 1______________________')
        dif_pai2_filho1 = [elemento for elemento in pai2 if elemento not in filho1]
        #print('_______________pai2_m1______________________')
        filho1_m1 = [12]
        filho1_m1.extend(seg1_ponto_pai1_m1)
        #print('_______________filho1_m1______________________')
        if len(filho1_m1)<len(pai2_m1)-1:
            while len(filho1_m1)<len(pai2_m1):
                filho1_m1.append(dif_pai2_filho1[0])
                dif_pai2_filho1.remove(dif_pai2_filho1[0])
            filho1_m1.append(12)
        else:
            filho1_m1.append(12)
        #print('________________filho1_m1______________________')
        pai2 = []
        pai2.extend(pai2_m1)
        pai2.extend(pai2_m2)
        #print('_______________Pai 2______________________')
        filho1 = []
        filho1.extend(seg1_ponto_pai1_m1)
        filho1.extend(seg1_ponto_pai1_m2)
        #print('_______________Filho 1______________________')
        dif_pai2_filho1 = [elemento for elemento in pai2 if elemento not in filho1]
        #print('_______________pai2_m2______________________')
        filho1_m2 = [11]
        filho1_m2.extend(seg1_ponto_pai1_m2)
        #print('_______________filho1_m2______________________')
        if dif_pai2_filho1 != []:
            while dif_pai2_filho1 != []:
                filho1_m2.append(dif_pai2_filho1[0])
                dif_pai2_filho1.remove(dif_pai2_filho1[0])
            filho1_m2.append(11)
        else:
            filho1_m2.append(11)
        #print('________________filho1_m1______________________')
        ## Filho2 M1
        pai2 = []
        pai2.extend(pai2_m1)
        pai2.extend(pai2_m2)
        #print('_______________Pai 2______________________')
        filho2 = [11]
        filho2.extend(seg2_ponto_pai1_m1)
        filho2.extend(seg2_ponto_pai1_m2)
        #print('_______________Filho 2______________________')
        dif_pai2_filho2 = [elemento for elemento in pai2 if elemento not in filho2]
        #print('_______________pai1_m1______________________')
        filho2_m1 = [12]
        filho2_m1.extend(seg2_ponto_pai1_m1)
        #print('_______________filho2_m1______________________')
        if len(filho2_m1)<len(pai1_m1)-1:
            while len(filho2_m1)<len(pai1_m1):
                filho2_m1.append(dif_pai2_filho2[0])
                dif_pai2_filho2.remove(dif_pai2_filho2[0])
            filho2_m1.append(12)
        else:
            filho2_m1.append(12)
        #print('________________filho2_m1______________________')
        #Crossover
        ## Filho2 M2
        pai2 = []
        pai2.extend(pai2_m1)
        pai2.extend(pai2_m2)
        #print('_______________Pai 2______________________')
        filho2 = [12]
        filho2.extend(seg2_ponto_pai1_m1[:])
        filho2.extend(seg2_ponto_pai1_m2[:])
        #print('_______________Filho 2______________________')
        filho2_m2 = [11]
        filho2_m2.extend(seg2_ponto_pai1_m2[:])
        #print('_______________filho2_m1______________________')
        if dif_pai2_filho2 != []:
            while dif_pai2_filho2 != []:
                filho2_m2.append(dif_pai2_filho2[0])
                dif_pai2_filho2.remove(dif_pai2_filho2[0])
            filho2_m2.append(11)
        else:
            filho2_m2.append(11)
        #print('________________filho2_m1______________________')
        ## Monto a matriz incial
        a_filhos = pd.read_excel('M1.xlsx')
        a_filhos = pd.DataFrame(a_filhos)
        i_filhos = list(range(1,jm+1))
        coluna_filhos = list(range(1,jm+1))

        a_filhos = a_filhos.set_axis(i_filhos, axis='index')
        a_filhos = a_filhos.set_axis(coluna_filhos, axis='columns')
        # Valores de Filho1 M1
        filho1_m1_valores = []
        i = 0
        while i < len(filho1_m1)-1:
            x = filho1_m1[i]
            y = filho1_m1[i+1]
            i = i+1
            a_filho1_m1 = a_filhos[y][x]
            filho1_m1_valores.append(a_filho1_m1)    
        # Valores de Filho1 M2
        filho1_m2_valores = []
        i = 0
        while i < len(filho1_m2)-1:
            x = filho1_m2[i]
            y = filho1_m2[i+1]
            i = i+1
            a_filho1_m2 = a_filhos[y][x]
            filho1_m2_valores.append(a_filho1_m2)    
        # Calculo do makespan do filho 1
        make_filho1 = []
        if np.sum(filho1_m1_valores) >= np.sum(filho1_m2_valores):
            make_filho1 = np.sum(filho1_m1_valores)
        else:
            make_filho1 = np.sum(filho1_m2_valores)
        # Valores de Filho2 M1
        filho2_m1_valores = []
        i = 0
        while i < len(filho2_m1)-1:
            x = filho2_m1[i]
            y = filho2_m1[i+1]
            i = i+1
            a_filho2_m1 = a_filhos[y][x]
            filho2_m1_valores.append(a_filho2_m1)    
        # Valores de Filho2 M2
        filho2_m2_valores = []
        i = 0
        while i < len(filho2_m2)-1:
            x = filho2_m2[i]
            y = filho2_m2[i+1]
            i = i+1
            a_filho2_m2 = a_filhos[y][x]
            filho2_m2_valores.append(a_filho2_m2)    
        # Calculo do makespan do filho 2
        make_filho2=[]
        if np.sum(filho2_m1_valores) >= np.sum(filho2_m2_valores):
            make_filho2 = np.sum(filho2_m1_valores)
        else:
            make_filho2 = np.sum(filho2_m2_valores)

        list_filho1_m1.append([filho1_m1])
        list_filho1_m2.append([filho1_m2])
        list_make_filho1.append(make_filho1)
        list_filho2_m1.append([filho2_m1])
        list_filho2_m2.append([filho2_m2])
        list_make_filho2.append(make_filho2)
        
                
        #list_filho2 = [filho2_m1,filho2_m2, make_filho2]
        #list_filho1 = np.extend(list_filho1)
        #list_filho2 = np.extend(list_filho2)
              
        gera = gera + 1
    list_filho1_m1.extend(list_filho2_m1)
    list_filho1_m2.extend(list_filho2_m2)
    list_make_filho1.extend(list_make_filho2)
    
    list_filho1_m1 = pd.DataFrame(list_filho1_m1)
    list_filho1_m1 = list_filho1_m1.set_axis(individuo, axis='index')
    m1_n = list(['m1'])
    list_filho1_m1 = list_filho1_m1.set_axis(m1_n, axis='columns')
    
    list_filho1_m2 = pd.DataFrame(list_filho1_m2)
    list_filho1_m2 = list_filho1_m2.set_axis(individuo, axis='index')
    m2_n = list(['m2'])
    list_filho1_m2 = list_filho1_m2.set_axis(m2_n, axis='columns')
    
    list_make_filho1 = pd.DataFrame(list_make_filho1)
    list_make_filho1 = list_make_filho1.set_axis(individuo, axis='index')
    make_n = list(['make'])
    list_make_filho1 = list_make_filho1.set_axis(m2_n, axis='columns')
    
    Populacao_seguinte = list_filho1_m1
    Populacao_seguinte['m2'] = list_filho1_m2
    Populacao_seguinte['make'] = list_make_filho1
    #display (list_filho2_m1, list_filho2_m2)
    display (Populacao_seguinte)
    print (Populacao_seguinte.to_latex())
    return pai1_m1, pai1_m2, apt_pai1, pai1_loc, pai2_m1, pai2_m2, apt_pai2, pai2_loc, filho1_m1, filho1_m2, make_filho1, filho2_m1, filho2_m2, make_filho2
```

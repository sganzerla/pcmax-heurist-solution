```python
####População inicial. 50 indivíduos criados aleatoriamente.
import numpy as np
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
import random
import matplotlib.pyplot as plt
def populacao_inicial(arq, n_pop):
    data = open(arq).readlines()
    #print(data)
    #-----------------machines ----------------------------#
    machines = int(data[0])
    # -----------------jobs ------------------------------#
    jobs = int(data[1])
     # --------------- times -------------------------------#
    # segunda linha em diante até linha 2 + quantidade de jobs
    str_times = ' '.join(str(i) for i in data[2: 4 + jobs]) # x[start:end:step]...x[:] means same as x[0:len(x):1]
    times = [int(i) for i in str_times.split() if i.isdigit()]
    #print ('Máquinas:',machines)
    #print ("Atividades:",jobs)
    #print ("Tempos:",times)
    jm = jobs + machines  # tarefas + maquinas dummys (linhas zeros)
    str_setups = ' '.join(str(i) for i in data[2 + jm: len(data) - 2 + jm])
    setups_list = [int(i) for i in str_setups.split() if i.isdigit()]
    setups_matrix = [setups_list[i::jm] for i in range(jm)]
    matriz_tempos = (pd.DataFrame(setups_matrix).T)
    t = times
    M1 = [[0 for j in range(jm)] for i in range(len(t))]
    M1[0][0] = matriz_tempos[0][0]
    for i in range(len(t)):
        for j in range(jm):
            if i == j:
                M1[i][j] = matriz_tempos[j][i] + 0
            else:
                M1[i][j] = matriz_tempos[j][i] + int(t[j])
    # matriz dos tempos
    a = pd.DataFrame(M1)
    i = list(range(1,jm+1))
    coluna = list(range(1,jm+1))
    a = a.set_axis(i, axis='index')
    a = a.set_axis(coluna, axis='columns')
    col = list(range(1,jm-1))
    jm = jobs + machines
    list_ind = []
    count = 0
    while count in range(n_pop): # gera a população com 50 indivíduos
        a = pd.DataFrame(M1)
        i = list(range(1,jm+1))
        coluna = list(range(1,jm+1))
        a = a.set_axis(i, axis='index')
        a = a.set_axis(coluna, axis='columns')
        col = list(range(1,jm-1))
        ibm = []
        icm= []
        icm = []
        ccm = []
        m1 = []
        m2 = []
        count = count + 1
        # Na linha 12 escolher um valor aleatório
        ibm = [12]
        cbm = []
        m1 = []
        y = jm
        x = random.choice(col)
        b = a[x][y]
        while b == 500:
            x = random.choice(col)
            b = a[x][y]
        a = a.drop(y, axis=0)
        a = a.drop(x, axis=1)
        #display ('del = ', a)
        i.remove(y)
        col.remove(x)
        y = x
        ibm.append(y)
        cbm.append(x)
        m1.append(b)
        # localizar a linha 12 e localizar o valor mínimo
        # Na linha 11 escolher um valor aleatório
        icm = [11]
        ccm = []
        m2 = []
        y = jm - 1
        x = random.choice(col)
        c = a[x][y]
        while c == 500:
            x = random.choice(col)
            c = a[x][y]
        a = a.drop(y, axis=0)
        a = a.drop(x, axis=1)    
        i.remove(y)
        col.remove(x)
        y = x
        icm.append(y)
        ccm.append(x)
        m2.append(c)
        while col != []:
            if np.sum(m1)<=np.sum(m2):
                y = ibm[-1]
                x = random.choice(col)
                b = a[x][y]
                while b == 500:
                    y = ibm[-1]
                    x = random.choice(col)
                    b = a[x][y]
                a = a.drop(y, axis=0)
                a = a.drop(x, axis=1)
                i.remove(y)
                col.remove(x)
                y = x
                ibm.append(y)
                cbm.append(x)
                m1.append(b)
            else:
                y = icm[-1]
                x = random.choice(col)
                c = a[x][y]
                while c == 500:
                    y = icm[-1]
                    x = random.choice(col)
                    c = a[x][y]
                a = a.drop(y, axis=0)
                a = a.drop(x, axis=1)
                i.remove(y)
                col.remove(x)
                y = x
                icm.append(y)
                ccm.append(x)
                m2.append(c)
        x = jm
        y = ibm[-1]
        b = a[x][y]
        ibm.append(x)
        m1.append(b)
        i.remove(y)
        a = a.drop(y, axis=0)
        a = a.drop(x, axis=1)
        x = jm - 1
        y = icm[-1]
        c = a[x][y]
        icm.append(x)
        m2.append(c)
        i.remove(y)
        if np.sum(m1) >= np.sum(m2):
            makespan = np.sum(m1)
        else:
            makespan = np.sum(m2)
        lista = [ibm,icm, makespan]
        list_ind.append(lista)
    #display (list_ind)
    return (M1, list_ind, jm)
```

```python
# Atribuir um peso f(i)=(1/make(i))/Somaj=1=>p(1/make(i))__________f(i)= inv_sum_make/sum_make
def ordem_prob(list_ind):
    ## Ordenação da lista População em ordem crescente de makespan
    # makespan = aptidão = fitness 
    # < makespan > aptidão
    individuo = list(range(1,51))
    list_ind = pd.DataFrame(list_ind)
    caractistica = list(['m1', 'm2', 'make'])
    #list_ind = list_ind.set_axis(individuo, axis='index')
    list_ind = list_ind.set_axis(caractistica, axis='columns')
    #display (list_ind)
    list_ind_crescente = list_ind.sort_values(by='make')
    list_ind_crescente = list_ind_crescente.set_axis(individuo, axis='index')
    #display(list_ind_crescente)
    make = list_ind_crescente['make']
    inv_sum_make = np.sum(1/make)
    list_prob = []
    i = 1
    while i in range(len(individuo)+1):
        f = (1/make[i])/(inv_sum_make)
        list_prob.append(f)
        i = i + 1
    list_prob = pd.DataFrame(list_prob)
    list_prob = list_prob.set_axis(individuo, axis='index')
    prob = list(['Probabilidade'])
    list_prob = list_prob.set_axis(prob, axis='columns')
    list_prob = pd.DataFrame(list_prob)
    list_ind_crescente_prob = list_ind_crescente.copy()
    list_ind_crescente_prob = pd.DataFrame(list_ind_crescente_prob)
    list_ind_crescente_prob['Probabilidade'] = list_prob
    list_ind_crescente_prob = list_ind_crescente_prob.sort_values(by='Probabilidade', ascending=False)
    list_ind_crescente_prob = list_ind_crescente_prob.set_axis(individuo, axis='index')
    #analise_pop_in = list_ind_crescente_prob.describe()
    #list_ind_crescente.hist(column="Probabilidade", bins=10)
    #plt.show()
    return (individuo, list_ind_crescente, list_ind_crescente_prob)
```

```python
populacao_inicial = populacao_inicial('instancia.txt', 50)
M1 = populacao_inicial[0]
list_ind = populacao_inicial[1]
jm = populacao_inicial[2]
```

```python
ordem_prob=ordem_prob(list_ind)
individuo = ordem_prob[0]
list_ind_crescente = ordem_prob[1]
list_ind_crescente_prob = ordem_prob[2]

display(list_ind_crescente_prob)
```
